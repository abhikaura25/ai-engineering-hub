```python
from langgraph.graph import StateGraph,START, END
from typing import TypedDict
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

load_dotenv()
model = ChatOpenAI()
```

```python
class BlogState(TypedDict):
    title: str
    outline: str
    content: str
    feedback: str

def generate_blog_outline(state: BlogState) -> BlogState:
    prompt = f"Generate a detailed blog outline for the following title: {state['title']}"
    outline = model.invoke(prompt)
    state['outline'] = outline.content
    return state

def generate_blog_content(state: BlogState) -> BlogState:
    prompt = f"Generate a detailed blog content based on the following outline: {state['outline']}"
    content = model.invoke(prompt)
    state['content'] = content.content
    return state

def evaluate_blog_quality(state: BlogState) -> BlogState:
    prompt = f"Evaluate the quality of the following blog content and provide feedback: {state['content']}"
    feedback = model.invoke(prompt)
    state['feedback'] = feedback.content
    return state
```

```python
graph = StateGraph(BlogState)

graph.add_node("generate_blog_outline", generate_blog_outline)
graph.add_node("generate_blog_content", generate_blog_content)
graph.add_node("evaluate_blog_quality", evaluate_blog_quality)

graph.add_edge(START, "generate_blog_outline")
graph.add_edge("generate_blog_outline", "generate_blog_content")
graph.add_edge("generate_blog_content", "evaluate_blog_quality")
graph.add_edge("evaluate_blog_quality", END)

compiled_graph = graph.compile()

result =compiled_graph.invoke({"title": "Pakistan Cricket"})

print(result['content'])

print("############# Feedback #############")
print(result['feedback'])
```
